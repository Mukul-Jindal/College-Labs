```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn import preprocessing
```

```python
AWsales = pd.read_csv('./archive/AWSales.csv')
```

```python
print(pd.isna(AWsales))
```

```python
AWCustomers = pd.read_csv('./archive/AWCustomers.csv')
print(pd.isna(AWCustomers))

```

```python
#Cleaning the data
AWCustomers=AWCustomers.drop(['Suffix','Title','FirstName','LastName'],axis=1)
AWCustomers
```

```python
AWCustomers['MiddleName'] = AWCustomers['MiddleName'].fillna(AWCustomers['MiddleName'].mode()[0])
```

```python
AWCustomers
```

```python
AWCustomers['CustomerID'].is_unique
```

```python
AWCustomers.drop_duplicates(subset='CustomerID',keep=False,inplace=True)
```

```python
AWCustomers['CustomerID'].is_unique
```

```python
AWCustomers['AddressLine2'] = AWCustomers['AddressLine2'].fillna(0)
```

```python
AWsales['CustomerID'].is_unique
```

```python
mergedDF = pd.merge(left=AWsales,right=AWCustomers,on='CustomerID')
```

```python
mergedDF
mergedDF.to_csv("AWsalesMergedAWcustomers.csv")
```

```python
scaler = preprocessing.MinMaxScaler()
AWCustomersCopy = mergedDF.copy()
AWCustomersCopy['YearlyIncome'] = scaler.fit_transform(np.array(AWCustomersCopy['YearlyIncome']).reshape(-1,1))
AWCustomersCopy['NumberCarsOwned'] = scaler.fit_transform(np.array(AWCustomersCopy['NumberCarsOwned']).reshape(-1,1))
```

```python
labelEncoder = preprocessing.LabelEncoder()
AWCustomersCopy['MaritalStatus']=labelEncoder.fit_transform(AWCustomersCopy['MaritalStatus'])
AWCustomersCopy['Gender']=labelEncoder.fit_transform(AWCustomersCopy['Gender'])
AWCustomersCopy
```

```python
bin_lables=['Low','Medium','High']
cut_bins=[0,0.33,0.66,1.0]
AWCustomersCopy['YearlyIncome']=pd.cut(AWCustomersCopy['YearlyIncome'],bins=cut_bins,labels=bin_lables)
AWCustomersCopy['NumberCarsOwned']=pd.cut(AWCustomersCopy['NumberCarsOwned'],bins=cut_bins,labels=bin_lables)
AWCustomersCopy
```
