```python
import pandas as pd
import numpy as np
```

Loading the Dataset

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
dataSet = pd.read_csv(url)
```

Adding Columns names to Dataset

```python
Col= ["symboling", "normalized_losses",
"make", "fuel_type", "aspiration","num_doors", "body_style", "drive_wheels",
"engine_location", "wheel_base", "length", "width", "height", "curb_weight",
"engine_type", "num_cylinders", "engine_size", "fuel_system", "bore", "stroke",
"compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]
dataSet.columns = Col
dataSet.head()
```

Replacing '?' values with NaN

```python
dataSet = dataSet.replace('?', np.nan)
dataSet
```

Replacing all NaN values with Central tendency and Droping NaN values in price columns

```python
dataSet  = dataSet.dropna(subset="price")
# dataSet.to_csv("test.csv")
# print(dataSet.isnull().sum())

col = ['normalized_losses','bore','stroke','horsepower','peak_rpm']
for i in col:
    median = dataSet[i].median()
    dataSet[i] = dataSet[i].fillna(median)

mode = dataSet['num_doors'].mode()
mode

dataSet['num_doors']=dataSet['num_doors'].fillna(mode[0])
dataSet[20:30]
```
