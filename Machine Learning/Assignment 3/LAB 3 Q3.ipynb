```python
import pandas as pd
import numpy as np
```

Loading the Dataset

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
dataSet = pd.read_csv(url)
```

Adding Columns names to Dataset

```python
Col= ["symboling", "normalized_losses",
"make", "fuel_type", "aspiration","num_doors", "body_style", "drive_wheels",
"engine_location", "wheel_base", "length", "width", "height", "curb_weight",
"engine_type", "num_cylinders", "engine_size", "fuel_system", "bore", "stroke",
"compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]
print(len(Col))
dataSet.columns = Col
dataSet.head()
```

Replacing '?' values with NaN

```python
dataSet = dataSet.replace('?', np.nan)
dataSet
```

Replacing all NaN values with Central tendency and Droping NaN values in price columns

```python
dataSet  = dataSet.dropna(subset="price")
# dataSet.to_csv("test.csv")
# print(dataSet.isnull().sum())

col = ['normalized_losses','bore','stroke','horsepower','peak_rpm']
for i in col:
    median = dataSet[i].median()
    dataSet[i] = dataSet[i].fillna(median)

mode = dataSet['num_doors'].mode()
mode

dataSet['num_doors']=dataSet['num_doors'].fillna(mode[0])
dataSet[20:30]
```

For “num_doors” and “num_cylinders”: convert words (number names) to figures
for e.g., two to 2

```python
print(dataSet['num_doors'].unique())
print(dataSet['num_cylinders'].unique())
stringToNumber = {
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'eight':8,
    'twelve': 12,
}
temp = dataSet.copy()
print(len(dataSet['num_doors']))

for key in stringToNumber.keys():
    temp['num_doors'] = temp['num_doors'].replace(key,stringToNumber[key])
    temp['num_cylinders'] = temp['num_cylinders'].replace(key,stringToNumber[key])
temp.head()
dataSet = temp.copy()
```

For "body_style", "drive_wheels": use dummy encoding scheme

```python
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder()
```

```python
dataSet = pd.get_dummies(dataSet,columns=['body_style','drive_wheels']).copy()
dataSet.head()
```

```python
dictTFValue={
    'True':1,
    'False':0,
}
colNames = ['body_style_convertible','body_style_hardtop','body_style_hatchback','body_style_sedan',	'body_style_wagon','drive_wheels_4wd','drive_wheels_fwd','drive_wheels_rwd']

for selectCol in colNames:
    dataSet[selectCol] = label_encoder.fit_transform(dataSet[selectCol])

dataSet.head()
```

For “make”, “aspiration”, “engine_location”,fuel_type: use label encoding
scheme

```python
colNames = ['make','aspiration','engine_location','fuel_type']
for col in colNames:
    dataSet[col]= label_encoder.fit_transform(dataSet[col])
dataSet.head()
```

For fuel_system: replace values containing string pfi to 1 else all values to 0.

```python
values = dataSet['fuel_system'].unique()
values

for val in values:
    if val != "spfi":
        dataSet['fuel_system'] = dataSet['fuel_system'].replace(val,0)

dataSet['fuel_system'] = dataSet['fuel_system'].replace('spfi',1)
```

For engine_type: replace values containing string ohc to 1 else all values to 0.

```python
values = dataSet['engine_type'].unique()

for val in values:
    if val != "ohc":
        dataSet['engine_type'] = dataSet['engine_type'].replace(val,0)

dataSet['engine_type'] = dataSet['engine_type'].replace('ohc',1)
```

Divide the dataset into input features (all columns except price) and output variable
(price). Scale all input features.

```python
X = dataSet.drop('price',axis='columns')
Y = dataSet['price']
X.head()
```

```python
scaler = preprocessing.MinMaxScaler()
colNames = ['normalized_losses','bore','stroke','compression_ratio','horsepower','peak_rpm','city_mpg','highway_mpg',"length", "width", "height", "curb_weight",'wheel_base']

for col in colNames:
    X[col] = scaler.fit_transform(np.array(X[col]).reshape(-1,1))

X.head()
```

Train a linear regressor on 70% of data (using inbuilt linear regression function of
Python) and test its performance on remaining 30% of data.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=42,test_size=0.20)
```

```python
model = LinearRegression()

model.fit(X_train,Y_train)
```

```python
prediction = model.predict(X_test)
print(prediction)
```

```python
import sklearn.metrics as sm 
```

```python
print("Mean absolute error =", round(sm.mean_absolute_error(Y_test, prediction), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(Y_test, prediction), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(Y_test, prediction), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(Y_test, prediction), 2)) 
print("R2 score =", round(sm.r2_score(Y_test, prediction), 2))
```

Reduce the dimensionality of the feature set using inbuilt PCA decomposition and then
again train a linear regressor on 70% of reduced data (using inbuilt linear regression
function of Python). Does it lead to any performance improvement on test set?

```python
from sklearn.decomposition import PCA 
```

```python
pca = PCA(n_components=10)
pca.fit(X)
pca_train_set= pca.transform(X_train)
pca_test_set= pca.transform(X_test)
```

```python
modelPCA = LinearRegression()
modelPCA.fit(pca_train_set,Y_train) 
```

```python
predictionPCA = modelPCA.predict(pca_test_set)
print(predictionPCA)
```

```python
print("Mean absolute error =", round(sm.mean_absolute_error(Y_test, predictionPCA), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(Y_test, predictionPCA), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(Y_test, predictionPCA), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(Y_test, predictionPCA), 2)) 
print("R2 score =", round(sm.r2_score(Y_test, predictionPCA), 2))
```
